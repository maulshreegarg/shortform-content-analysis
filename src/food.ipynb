```python
file_path = '/content/food_preprocessed.xlsx'
import pandas as pd
df = pd.read_excel(file_path)
print(df)
```

```python
print(df.head())
```

```python
df.info()
```

```python

df.drop_duplicates(inplace=True)
```

```python
df.fillna(method='ffill', inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='title', y='ytlikes', hue='ytlikes')
plt.xlabel('Title')
plt.ylabel('Number of Likes')
plt.plot(df['title'], df['ytlikes'], linestyle='-', marker='o', color='r', label='Likes')
plt.plot(df['title'], df['instalikes'], linestyle='-', marker='o', color='b', label='Views')
plt.show()
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(data=df, x='duration', y='ytlikes', hue='ytlikes',width=2)
plt.title('Bar Plot ')
plt.show()
```
